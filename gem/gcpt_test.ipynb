## Using GEM data on operating plants

```python
import pandas as pd
df = pd.read_excel("data/GCPT_H12025.xlsx")
df.head()
```

```python
df.columns
```

```python
df = df.rename(columns={'Annual CO2 (million tonnes / annum)': 'Annual CO2', 
                                'Lifetime CO2 (million tonnes)': 'Lifetime CO2',
                                'Heat rate (Btu per kWh)': 'Heat rate',
                                'Emission factor (kg of CO2 per TJ)': 'Emission factor',
                                'Plant age (years)': 'Plant age',
                                'Remaining plant lifetime (years)': 'Remaining plant lifetime'})
```

```python
#filter by status
data = df[df["Status"].isin(["operating", "mothballed"])]
data["Status"].value_counts()
```

```python
#filter columns
test_columns = ["Capacity (MW)", "Longitude", "Latitude", "Plant age", "Heat rate",
                "Emission factor", "Annual CO2", "Remaining plant lifetime", "Lifetime CO2"]

subset = data[test_columns].copy()
subset = subset.dropna()
```

```python
#do a umap projection

import umap
reducer = umap.UMAP(random_state=42)
embedding = reducer.fit_transform(subset)
```

```python
data["umap_x"] = embedding[:, 0]
data["umap_y"] = embedding[:, 1]
```

```python
data["umap_x"]
```

```python
import networkx as nx
from sklearn.neighbors import NearestNeighbors
import umap

# 1. Take a subset of rows and just the feature columns for UMAP
rows = data.iloc[:1000].copy()
X = rows[test_columns].copy()  

# 2. UMAP embedding
reducer = umap.UMAP(random_state=42)
embedding = reducer.fit_transform(X)  # shape: (n_samples, 2)

# Attach embedding to the same dataframe 
rows["umap_x"] = embedding[:, 0]
rows["umap_y"] = embedding[:, 1]

# 3. Build a k-NN graph using the embedding
knn = NearestNeighbors(n_neighbors=8).fit(embedding)
distances, indices = knn.kneighbors(embedding)

G = nx.Graph()

# 4. Add nodes (one per row/unit)
for idx, row in rows.iterrows():
    node_id = row["Unit ID"] if "Unit ID" in rows.columns else idx
    G.add_node(
        node_id,
        capacity=row.get("Capacity (MW)"),       
        lat=row.get("Latitude"),
        lon=row.get("Longitude"),
        age=row.get("Plant age"),
        tech=row.get("Heat rate"),
        coal=row.get("Emission factor"),
        ann_emissions=row.get("Annual CO2"),
        life_emissions=row.get("Lifetime CO2"),
        life=row.get("Remaining plant lifetime"),
        x=row["umap_x"],
        y=row["umap_y"],
    )

# Map from integer position in `embedding` back to node ids
node_ids = list(rows["Unit ID"] if "Unit ID" in rows.columns else rows.index)

# 5. Add edges based on k-NN indices
for i, neighbors in enumerate(indices):
    src = node_ids[i]
    for j in neighbors:
        if i == j:
            continue
        dst = node_ids[j]
        # Optional: weight by distance
        G.add_edge(src, dst, weight=distances[i][list(neighbors).index(j)])

# 6. Explore / visualize
explorer = Explore(G=G, raw_df=None)  
```

```python
explorer.drawGraph()
```
